```python
import matplotlib.pyplot as plt
from random import *
import numpy as np
import mpld3

x = range(0, 1000)
y = []

for n in x:
    y.append(randint(0,50))

#Linear
fit1 = np.polyfit(x, y, 1 , full = True)
y_1 = np.polyval(fit1[0],x)


#Quadratic
fit2 = np.polyfit(x, y, 2 , full = True)
y_2 = np.polyval(fit2[0],x)


#Cubic
fit3 = np.polyfit(x, y, 3 , full = True)
y_3 = np.polyval(fit3[0],x)


#Quartic
fit4 = np.polyfit(x, y, 4 , full = True)
y_4 = np.polyval(fit4[0],x)


#Fifth Degree
fit5 = np.polyfit(x, y, 5 , full = True)
y_5 = np.polyval(fit5[0],x)

plt.figure(figsize = (15,15))
plt.plot(x,y,"bo", label="Random Noise")
plt.plot(x, y_1, "g", label="Linear")
plt.plot(x,y_2, "y", label = "Quadratic")
plt.plot(x,y_3, "r", label = "Cubic")
plt.plot(x, y_4, "k", label = "Quartic")
plt.plot(x, y_5, "c", label = "Fifth Degree")
plt.legend()
plt.show()

print("The residuals for the linear fitting is", fit1[1])
print("The residuals for the quadratic fitting is", fit2[1])
print("The residuals for the cubic fitting is", fit3[1])
print("The residuals for the quartic fitting is", fit4[1])
print("The residuals for the fifth degree fitting is", fit5[1])
```

Clearly, the fifth degree provides the best fit since it has a smaller residual.
