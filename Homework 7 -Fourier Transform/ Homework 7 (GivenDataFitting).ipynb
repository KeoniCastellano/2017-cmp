```python
#Graph of fourier coefficients for the data
c = fft.fft(data)
plt.figure(figsize=(15,15))
plt.plot(x,np.abs(c),'-r')
plt.show()
```

```python
#Looks like we can take coefficients whose magnitude is bigger than 1000

c2 = c[np.abs(c)>1000]
y1 = fft.ifft(c2)/len(c2)

x1 = np.linspace(0,1000,len(y1))

plt.semilogx(x1,y1.real,label="FT")
plt.semilogx(x,data, label='Original')
plt.legend()
plt.plot()
```
