```python
'''
2. Non-uniform random number generator
'''

# To generate the exponential distribution from uniform distribution
import numpy as np
import matplotlib.pyplot as plt

tau = 3.053*60
N = 1000
mu = np.log(2)/tau
rand = []
x = np.linspace(0,1,1000)

for i in range(N):    
    z0 = np.random.random()
    x0 = int(-np.log(1-z0)/mu) #the time to decay from random number
    rand.append(x0)
        
plt.plot(x, rand)


plt.show()
```

```python
'''
Here I use the numpy function for the exponentially distributed random numbers to check if I did things right
Looks like I did...I think
'''
y = []
for i in range(1000):
    y.append(np.random.exponential(scale = mu))
plt.plot(x, y)
plt.show()
```
