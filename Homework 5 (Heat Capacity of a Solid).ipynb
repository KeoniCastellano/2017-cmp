```python
from scipy.integrate import quad 
from numpy import exp 
import matplotlib.pyplot as plt


#Constants
dens = 6.022 * (10 ** 28)   #m^3
DebTemp = 428               #K
V = 1000 * (10**-6)         #m^3
k = 1.381 * (10**-23)       #J/K


#Function in the integral for the Debye heat capacity
f = lambda x: ((x**4)*(exp(x)))/((exp(x)-1)**2)


#Formula to calculate heat capacity for a given temperature
C = lambda T: 9 * V * dens * k * ((T/DebTemp) ** 3) * quad(f, 0 , DebTemp/T)[0]



heatcap = float(input("Please enter a temperature in Kelvin: "))
print("The heat capacity for temperature {:.2f} K is {:.2f} J/K.".format(heatcap, C(heatcap)))
```

```python
#WE GRAPH HEAT CAPACITY VS TEMPERATURE USING THE DEBYE MODEL OF HEAT CAPACITY
Temps = range(5, 500, 1)
HeatCaps = []
for t in Temps:
    HeatCaps.append(C(t))

plt.figure(figsize=(15,15))
plt.plot(Temps,HeatCaps)
plt.xlabel("Temperature(K)", size = 36)
plt.ylabel("$C_V$ (J/K)", size = 36)
plt.title("Heat Capacity vs Temperature", size = 36)
plt.show()
```
