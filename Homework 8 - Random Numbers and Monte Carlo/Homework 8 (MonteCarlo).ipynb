```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from numpy import exp, pi,sqrt
from scipy import integrate
from random import random

f = lambda x: (x**(-1/2))/(exp(x)+1)
```

```python
'''
We first graph the function to determine the region that we are looking at.
'''
x = np.linspace(0.0001,1,1000)
y = f(x)

plt.figure(figsize=(15,15))
plt.title('Graph of Function')
plt.axis([0,1,0,1])
plt.plot(x,y)
plt.show()
```

We now implement the importance sampling method of Monte Carlo integration using $w(x)=x^{-1/2}$ and 1,000,000 sampling points.

```python
'''
This is the algorithm for importance sampling
'''
w = lambda x: x**(-1/2)
const = integrate.quad(w,0,1)[0]
def IS(N):
    ratio = 0
    for i in range(N):
        x = random()
        ratio += (f(x)/w(x))
    I = ratio*const/N
    return I
```

```python
print('The value of the integral using importance sampling is %.4f' %IS(1000000))
```
