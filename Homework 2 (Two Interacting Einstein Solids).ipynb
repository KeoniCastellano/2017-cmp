```python
#Keoni Castellano
#Homework 2
#Two Interacting Einstein Solids

#Part 1
#This program will calculate the total number of microstates of a system of two weakly coupled Einstein solids A and B given
#only q_A, the number of energy units in A.
#We will calculate the multiplicity when N_A=300, N_B=200, and q=100.


#Input: An integer greater than 0 and less than q=100
#Output: The multiplicity of the specified system


#Constants
NA = 300
NB = 200
q = 100

#Program start
import numpy as np
import matplotlib.pyplot as plt
import math


#This function takes as input two positive integers n and k, checks that n is greater than or equal to k, and returns
#the binomial coefficient bino(n,k) or n choose k.
#Input: Two positive integers
#Output: Positive integer

def bino(n, k):
    "We need to make sure that n is greater than or equal to k."
    if n >= k:
        num = math.factorial(n)
        denom = math.factorial(k) * math.factorial(n-k)
        return num//denom
    if n < k:
        return 0


qA = int(input("Please enter an integer greater than 0 and less than 100. This will be the number of energy units in A: "))

while qA < 0 or qA > 100:
    print("That is invalid!!!")
    qA = int(input("Please enter an integer greater than 0 and less than 100. This will be the number of energy units in A: "))

qB = q - qA

omA = bino(qA + NA - 1, qA)
omB = bino(qB + NB - 1, qB)


#Output
print("")
print("The system that we have is: ")
print("")
print("q(A) = ", qA)
print("")
print("q(B) = ", qB)
print("")
print("o(A) = ", omA)
print("")
print("o(B) = ", omB)
print("")
print("o(total) = ", omA * omB)
```

```python
#Plot for Part 1
#Here we graph Omega(total) as a function of qA.
x = range(0, 101)
y = []


for i in x:
    qA = i
    qB = q - qA
    omA = bino(qA + NA - 1, qA)
    omB = bino(qB + NB - 1, qB)
    y.append(omA * omB)
    

plt.plot(x,y)
plt.title("$N_A$ = 300, $N_B$ = 200, q = 100")
plt.xlabel("$q_A$")
plt.ylabel("$\Omega_total$")
plt.show()
```

```python
#Here we will plot a graph of O(total) vs qA for q = 100 to see the general trend as N gets larger and larger.
#First we define a function that will take a number for N and plot Omega(Total).

def Omega(N):
    x = range(101)
    y = []
    NA = 2*(math.exp(math.log(N)-math.log(5)))
    NB = N - NA
    q = 100

    for i in x:
        qA = i
        qB = q - qA
        omA = bino(int(qA) + int(NA) - 1, int(qA))
        omB = bino(int(qB) + int(NB) - 1, int(qB))
        y.append(omA * omB)


    plt.plot(x,y)
    plt.title("N = " + str(N))
    plt.xlabel("$q_A$")
    plt.ylabel("$\Omega_total$")
    plt.show()
```
